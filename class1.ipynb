### 1. Relu

Relu (Rectified Linear Unit) 

Works in the following way:
* If the input value is less than 0, return 0.
* If the input value is greater than or equal to 0, return the input value.

**Use cases:**
Because of its non-linearity allows neural networks to learn complex patterns and making backpropagation more efficient.

The mathematical representation of this is:
$$
f(x) = \max(0, x)
$$

### 2. Sigmoid

**Use cases**
* It ensures a smooth and continuous output that is essential for gradient-based optimization methods.
* It also allows neural networks to handle complex patterns that linear equations cannot. 

It is mathematically defined as
$$
f(x) = \frac{1}{1 + e^{-x}}
$$

### 3. Tanh (Hyperbolic Tangent)

Tanh (hyperbolic tangent function), is similar to the sigmoid function with the difference that it stretches across the y-axis. 

**Use cases:**
* Commonly used in hidden layers due to its zero-centered output, facilitating easier learning for subsequent layers.

It is defined as: 

$$
f(x) = \tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$

### 4. Leaky ReLU

Works as an improved version of ReLU function to solve the Dying ReLU problem as it has a small positive slope in the negative area.

**Use cases**
* same as the ones of ReLU, but adding that it does enable backpropagation, even for negative input values.

$$
f(x) =
\begin{cases}
x & \text{if } x \geq 0 \\
\alpha x & \text{if } x < 0
\end{cases}
$$

### 5. Softmax function

Non-linear activation function that allows to transform the output scores from a neural network into probabilities by squashing the output values of each class into the range of 0 to 1, and ensuring that the sum of all probabilities equals 1. 

**Use cases**
* Useful to handle multi-class classification problems

$$
f(x_i) = \frac{e^{x_i}}{\sum_{j=1}^{n} e^{x_j}}, \quad i = 1, 2, \ldots, n
$$

## Activity 2
For a given X values.
Plot the output from 5 activation functions.
Relu, Sigmoid and others

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_sigmoid():
    x = np.linspace(-10, 10, 100)  # Generate 100 equally spaced values from -10 to 10
    y = 1 / (1 + np.exp(-x))  # Compute the sigmoid function values
    
    plt.plot(x, y)
    plt.xlabel('Input')
    plt.ylabel('Sigmoid Output')
    plt.title('Sigmoid Activation Function')
    plt.grid(True)
    plt.show()
    
plot_sigmoid()
```

```python
def plot_tanh():
    # Generate values for x
    x = np.linspace(-10, 10, 100)

    # Compute tanh values for corresponding x
    tanh = np.tanh(x)

    # Plot the tanh function
    plt.plot(x, tanh)
    plt.title("Hyperbolic Tangent (tanh) Activation Function")
    plt.xlabel("x")
    plt.ylabel("tanh(x)")
    plt.grid(True)
    plt.show()

plot_tanh()
```

```python
def plot_relu():
    # Generate values for x
    x = np.linspace(-10, 10, 100)

    # Compute ReLU values for corresponding x
    relu = np.maximum(0, x)

    # Plot the ReLU function
    plt.plot(x, relu)
    plt.title("ReLU Activation Function")
    plt.xlabel("x")
    plt.ylabel("ReLU(x)")
    plt.grid(True)
    plt.show()
    
plot_relu()
```

```python
def plot_leaky_relu():
    # Generate values for x
    x = np.linspace(-10, 10, 100)

    # Define the leaky ReLU function
    def leaky_relu(x, alpha=0.1):
        return np.where(x >= 0, x, alpha * x)

    # Compute leaky ReLU values for corresponding x
    leaky_relu_values = leaky_relu(x)

    # Plot the leaky ReLU function
    plt.plot(x, leaky_relu_values)
    plt.title("Leaky ReLU Activation Function")
    plt.xlabel("x")
    plt.ylabel("Leaky ReLU(x)")
    plt.grid(True)
    plt.show()
    
plot_leaky_relu()
```

```python
def plot_softmax(probabilities, class_labels):
    plt.bar(class_labels, probabilities)
    plt.xlabel("Class")
    plt.ylabel("Probability")
    plt.title("Softmax Output")
    plt.show()

# Example usage:
class_labels = ["Class A", "Class B", "Class C"]
probabilities = np.array([0.2, 0.3, 0.5])
plot_softmax(probabilities, class_labels)
```
